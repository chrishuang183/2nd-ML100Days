```python
import requests
response = requests.get(target_url)
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[0:100]
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = "\t"

data = data.split(split_tag)
print(len(data))
data[0]
```

將 txt 轉成 pandas dataframe

```python
import pandas as pd

    
df = pd.DataFrame(data)
df.head()
```
